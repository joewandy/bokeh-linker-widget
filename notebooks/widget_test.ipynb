```python
import numpy as np
```

```python
from random import random

from bokeh.layouts import row
from bokeh.models import CustomJS, ColumnDataSource
from bokeh.models.widgets import DataTable, TableColumn
from bokeh.plotting import figure, output_notebook, show

output_notebook()

label = 'spectra'
x1 = ['%s_%d' % (label, i) for i in list(range(100))]
y1 = np.random.randint(low=1, high=100, size=100)

label = 'MF'
x2 = ['%s_%d' % (label, i) for i in list(range(100))]
y2 = np.random.randint(low=1, high=100, size=100)

label = 'BGC'
x3 = ['%s_%d' % (label, i) for i in list(range(100))]
y3 = np.random.randint(low=1, high=100, size=100)

label = 'GCF'
x4 = ['%s_%d' % (label, i) for i in list(range(100))]
y4 = np.random.randint(low=1, high=100, size=100)

s1 = ColumnDataSource(data=dict(x=x1, y=y1))
s2 = ColumnDataSource(data=dict(x=x2, y=y2))
s3 = ColumnDataSource(data=dict(x=x3, y=y3))
s4 = ColumnDataSource(data=dict(x=x4, y=y4))

s1.selected.js_on_change('indices', CustomJS(args=dict(s1=s1, s2=s2, s3=s3, s4=s4), code="""
        var s1_indices = cb_obj.indices;
        var new_s2_indices = [];
        var new_s3_indices = [];
        var new_s4_indices = [];
        
        for (var i = 0; i < s1_indices.length; i++) {
            var idx = s1_indices[i];
            new_s2_indices.push(idx+1);
            new_s3_indices.push(idx+2);
            new_s4_indices.push(idx+3);            
        }
        s2.selected.indices = new_s2_indices;
        s3.selected.indices = new_s3_indices;
        s4.selected.indices = new_s4_indices;

        // example to change data
        // var d1 = s1.data;
        // var d2 = s2.data;
        // for (var i = 0; i < inds.length; i++) {
        //    d2['x'].push(d1['x'][inds[i]]+100)
        //    d2['y'].push(d1['y'][inds[i]]+100)
        // }        
        
        s2.change.emit();
        s3.change.emit();
        s4.change.emit();
        
    """)
)

columns = [
        TableColumn(field="x", title="x"),
        TableColumn(field="y", title="y"),
    ]
data_table_1 = DataTable(source=s1, columns=columns, width=300, height=300)
data_table_2 = DataTable(source=s2, columns=columns, width=300, height=300)
data_table_3 = DataTable(source=s3, columns=columns, width=300, height=300)
data_table_4 = DataTable(source=s4, columns=columns, width=300, height=300)

layout = row(data_table_1, data_table_2, data_table_3, data_table_4)
show(layout)
```
